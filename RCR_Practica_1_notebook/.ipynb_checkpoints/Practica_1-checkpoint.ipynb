# Práctica 1 URDF

## Objetivo

El objetivo de esta práctica es que el alumno comprenda, interprete y modifique la información básica de los formatos de Universal Robot Description File (URDF).


### Metas 

- Que el alumno comprende el orden de los archivos que comprenden los espacios de trabajo (ws) y los paquetes que los integran en ROS 2.
- Que el alumno comprenda la estructura básica de un archivo URDF.
- Que el alumno sea capaz de mofificar la información básica de los archivos URDF para modificar la descripción de un robot serial.
- Que el alumno conozca los elementos de los archivos "launch" en ROS 2.
- Que el alumno sea capaz de vizualizar un robot descrito en un archivo URDF mediante la aplicación de RVIZ.

## Previo de la Práctica 1

Se le solicita al alumno investigar y colocar la siguiente información en esta sección:

1. Definición de URDF.

    **URDF** son las siglas para "_Universal Robot Description Format_" [1] o "_Unified Robot Description Format_" [2]. Es un paquete que contiene las especificaciones del robot en XML. Estas especificaciones abarcan el modelo del robot (geometría y organización), sensores, escenas, entre otras características dentro de ROS. A cada especificación XML corresponde un _parser_; es decir, un programa que toma texto y lo convierte en estructuras de datos. Este _parser_ sirve como traductor para uno o más lenguajes.

2. Tipos de juntas que están definidas dentro del formato **URDF** [3].
 
    - _Revoluta_: Es una junta que permite la rotación alrededor de un eje con límites máximos y mínimos establecidos.
      
    - _Continua_: Es una junta similar a la revoluta, pero no tiene límites por lo que permite una rotación continua.
 
      
    - _Prismática_: Es una junta que permite un movimiento lineal de deslizamiento. Posee un rango de movimiento con límites superior e inferior.
 
      
    - _Fija_: No es propiamente una junta. Bloquea el movimiento y no posee grados de libertad._Flotante_: Permite el movimiento en los 6 grados de libertad.
 
      
    - _Plana_: Esta junta permite el movimiento en el plano perpendicular a un eje establecido.

3. Definición de elementos "primitivos" que se pueden utilizar para definir un robot en ROS [4].

   Los elementos primitivos que se pueden utilizar para definir un robot son 3:

      - _Box_: Requiere que se especifiquen sus tres dimensiones en el atributo **size**.
      - _Cylinder_: Requiere que se especifiquen su radio y largo.
      - _Sphere_: Únicamente se debe especificar su radio para definirla.

        En todos los casos se establecerá el origen de su marco de referencia local en su centro.

        Adicionalmente, se encuentra la clasificación _mesh_, por medio de la cual se pueden importar figuras más complejas. Únicamente se requiere especificar el nombre del archivo con su ruta respectiva y un factor de escala. Cualquier formato de geometría se acepta.

4. Qué es un nodo y que es un tópico en ROS 2

    Un nodo en ROS 2 es el elemento encargado de realizar una tarea de forma específica y modular. Es capa de recibir y enviar información a través de tópicos, servicios, acciones o parámetros. De manera práctica estos nodos son programas que están ejecutándose en forma paralela con una única función. Algunos ejemplos de funciones para los nodos son el control de motores, recepción de datos de sensores, entre otras. [5]

  Por otro lado, los tópicos son el medio por el cual dos nodos inercambian información por medio de mensajes. Actúan como bus o canal de comunicación. Un nodo no está limitado en el número de tópicos a los que puede estar suscrito ni a la cantidad de mensajes que puede publicar de manera simultanea. [6]

5. Qué es RVIZ.

    RViz es un visualizador 3D para el entorno de ROS (_Robot Operating System_) [7], mantenido por la _Open Source Robotics Foundation_. Su nombre proviene de la contracción de _ROS Visualization_. RViz tiene un papel fundamental para el desarrollo y simulación dentro del ambiente de ROS ya que, como su nombre lo indica, provee una interfaz gráfica que permite visualizar diversos aspectos de la operación de un robot. Brinda una ruta para el entendimiento de los datos provenientes de sensores, la planeación de trayectorias y las transformaciones [8].

## Desarrollo de la práctica 

### Configuración del espacio de trabajo

En primer lugar, es necesario crear el espacio de trabajo donde se creará el paquete para el URDF. Para el caso de esta práctica el espacio de trabajo tiene por nombre `robot_2025_ws`. Al momento de crear el directorio del espacio de trabajo, en su interior se creó el directorio `src` para almecenar todos nuestros directorios y archivos de desarrollo. Para finalizar, por medio de la instrucción `colcon build` se debe compilar al directorio como _workspace_ de ROS.

Dentro de la carpeta `src` se construye el paquete de ROS al ejecutar el comando:

>```
> $ ros2 pkg create --build-type ament_cmake practica1_description
>```

Aquí el argumento `ament_cmake`indica que se va a emplear un archivo `CMakeLists.txt` para especificar los parámetros de compilación. El segundo argumento indica el nombre que tendrá el paquete; en este caso, `practica1_description`. Dentro del directorio del paquete se crearán tres nuevos directorios: `urdf`, `rviz` y `launch`, adicionales a los archivos y directorios creados al construir el paquete.

Posteriormente es necesario agregar la función que se muestra dentro del archivo `CMakeLists.txt` antes del llamado a función `ament_package()`. Aquí se especifica qué directorios serán compartidos para el resto de elementos del paquete.

```c
install(
    DIRECTORY launch rviz urdf
    DESTINATION share/${PROJECT_NAME}/
)
```
En el archivo `package.xml` se incluye información relevante sobre el paquete. Se incluye la versión, una descripción, datos de contacto de la persona encargada del mantenimiento y la licencia. Además, se especifica el constructor empleado `ament_cmake`.

### Archivo URDF

Dentro del directorio `urdf` ejecuta la instrucción `$ touch basic_link.urdf` para crear el archivo con las especificaciones en HTML del robot. Dentro de este archivo se comienza a definir el robot por el eslabón básico. Se deben definir su aspecto visual, sus límites de colisión y sus características inerciales, como la masa y su matriz de inercia.

Dentro del directorio `launch` se crea el archivo `basic_link_display.launch.xml`donde se declaran las rutas del archivo **URDF** y el archivo de configuración de RViz. Además, se declaran los nodos que van a componer el programa: el publicador del estado del robot `robot_state_publisher`, el publicador del estado de las juntas que servirá para interactuar con el robot `joint_state_publisher_gui`y el que corresponde a RViz donde podremos visualizar las juntas y eslabones `rviz2`.

Antes de construir y ejecutar el paquete es necesario instalar **Xacro**. Este es un lenguaje de macros que permite limpiar el código de HTML antes de convertirlo a una serie de instrucciones en ensamblador. Además, instalamos la interfaz gráfica de usuario que permitirá actualizr el estado de las juntas. Esto se realiza mediante el siguiente bloque de instrucciones en terminal. 

> ```
> # apt-get update
> # apt install ros-humble-xacro
> # apt install ros-humble-joint-state-publisher-gui
> ```

Hecho esto, basta con ejecutar el siguiente bloque de instrucciones que se muestra a continuación dentro del directorio `practica1_description` para construir y ejecutar el paquete de ROS que nos permitirá visualizar el robot.

> ```
$ colcon build
$ source install/setup.bash
$ ros2 launch practica1_description basic_link_display.launch.xml
> ```

### Configurar el ambiente de RViz

Al ejecutar la última instrucción de la sección anterior, si no hay ningún error, se inicia el ambiente de RViz. Sin embargo, falta realizar algunos ajustes para poder visualizar e interactuar con el robot. En primer lugar, se debe seleccionar el eslabón base _base_link_ como el marco de referencia fijo. Posteriormente se deben crear las visualizaciones _RobotModel_ y _TF_. Dentro de la primera se debe especificar que el tópico con la descripción se llama _/robot_description_ y asegurarse de que la casilla _Visual Enabled_ esté marcada. Opcionalmente se pueden habilitar los marcos de referencia y sus nombres.

Hecho esto se puede comenzar a interctuar con el modelo.

### Añadir eslabones al modelo

Hecha la configuración del paquete y del archivo **URDF**, es posible comenzar a interactuar y a modificar las características del modelo de robot únicamente a través del archivo `basic_link.urdf`. Únicamente es necesario tomar en cuenta que por cada eslabón creado posterior al eslabón base es necesario agregar y describir la junta por medio de la cual va unida al resto del robot.

En el desarrollo de esta práctica se declararon dos eslabones, además del eslabón base, con la primitiva _box_. Todos los eslabones se especificaron con las mismas dimensiones: una base de 0.2 x 0.2 [m] y con una altura de 0.8 [m]. Todos se declararon con masa unitaria y una matriz de inercia simétrica con los momentos de inercia unitarios. El eslabón base es rojo, el primer eslabón es verde y el segundo es azul mediante el código RGBA, que permite modificar la transparencia del color seleccionado con su última componente. Las características del eslabón se especifican de la siguiente forma:

```xml
<link name="base_link">
    <visual>
        <origin xyz="0.0 0.0 0.4353" rpy="0.0 0.0 0.0"/>
        
        <geometry>
            <box size="0.2 0.2 0.8"/>
        </geometry>

        <material name="color_1">
            <color rgba="1.0 0.0 0.0 1.0"/>
        </material>
    </visual>
    
    <collision>
        <origin xyz="0.0 0.0 0.4353" rpy="0.0 0.0 0.0"/>
        
        <geometry>
            <box size="0.2 0.2 0.8"/>
        </geometry>
    </collision>
    
    <inertial>
        <mass value="1.0"/>

        <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
        
        <inertia ixx="1.0" ixy="0.0" ixz="0.0"
            iyy="1.0" iyz="0.0" izz="1.0"/>
    </inertial>
</link>
```

Adicionalmente, debido al número de eslabones, se declararon dos juntas, ambas revolutas con un rango de movimiento de $-\frac{\pi}{4}$ a $\frac{\pi}{4}$. En la primera se declara al eslabón base como eslabón padre y al primer eslabón como eslabón hijo. Ésta gira alrededor del eje X. En la segunda junta el primer eslabón actúa como eslabón padre y el segundo eslabón como hijo. Éste gira alrededor del eje Y. La descripción de una junta se muestra a continuación:

```xml
<joint name="base_link_to_link1_joint" type="revolute" >
    <parent link="base_link" />
    <child link="link_1" />
    <axis xyz="1.0 0.0 0.0" />
    <limit lower="-0.78" upper="0.78" velocity="30"
        effort="10" />
    <origin xyz="0.0 0.0 0.8707" rpy="0.0 0.0 0.0" />
</joint>
```

## Resultados

Habiendo llevado a cabo el proceso descrito en el desarrollo de la práctica, se pudo modelar, simular e interactuar con un robot serial de dos eslabones móviles. En la figura 4, se puede visualizar el resultado en el ambiente del simulador RViz . Para lograr que los sistemas relativos a cada eslabón no estuvieran en el centro del cuerpo se colocó el origen visual, de colisión e inercial con un desplazamiento positivo de la mitad de la altura sobre el eje Z. Además, se consideró la distancia de separación necesaria para evitar coalisiones en el movimiento rotacional del eslabón como $\frac{L_x}{2} \cdot \sin(\theta_{máx}) = 0.1 \cdot \sin(\frac{\pi}{4})  = 0.0707$. De esta forma, el origen de los eslabones se estableció en $[0.0\; 0.0\; 4.353]$ y el eje de las juntas en $[0.0\; 0.0\; 0.8707]$. Así, además de conseguir un mejor efecto visual, permite que el movimiento de las juntas sea mucho más suave.
